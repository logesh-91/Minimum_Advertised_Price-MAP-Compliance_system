```python
import pandas as pd

PROMOTIONTABLE = pd.read_csv('promotion table.csv')

def get_promotions_by_quarter(target_quarter):

    valid_quarters = ['Q1', 'Q2', 'Q3', 'Q4']
    if target_quarter not in valid_quarters:
        return f"Error: Please choose from {valid_quarters}"

    result_df = PROMOTIONTABLE[PROMOTIONTABLE['session'].str.startswith(target_quarter)].copy()
    
    result_df['Selected_Quarter'] = target_quarter
    
    return result_df

df = get_promotions_by_quarter('Q1')

if isinstance(df, str):
    print(df)
else:
    print("Promotions for Q1:")

    print(df[['SKU', 'PL', 'session', 'promotion']].head(10))
```

```python
import pandas as pd
import re

def calculate_lpp(map_price, promo_str):                     
   
    
    if pd.isna(promo_str) or pd.isna(map_price):
        return map_price
    
    promo_str = str(promo_str).strip()
    
    perc_match = re.search(r'(\d+)\s*%', promo_str)
    if perc_match:
        percentage = float(perc_match.group(1)) / 100.0
        return map_price * (1 - percentage)
    
    fixed_match = re.search(r'\$\s*(\d+)', promo_str)
    if fixed_match:
        discount_amount = float(fixed_match.group(1))
        return map_price - discount_amount
    

    if "Free" in promo_str:
        return map_price
    
    return map_price

def run_lpp_process(csv_path, yaml_path):

    df_sku = pd.read_csv(csv_path)


    with open(yaml_path, 'r') as f:
        yaml_content = f.read()
    
    entries = re.findall(r'MAP:\s*(\d+).*?sku:\s*(\w+)', yaml_content, re.DOTALL)
    df_map = pd.DataFrame(entries, columns=['MAP', 'SKU'])
    df_map['MAP'] = df_map['MAP'].astype(float)


    df_final = pd.merge(df_sku, df_map, on='SKU', how='left')


    quarters = ['Q1', 'Q2', 'Q3', 'Q4']
    for q in quarters:
        lpp_column_name = f'LPP_{q}'
        df_final[lpp_column_name] = df_final.apply(
            lambda row: calculate_lpp(row['MAP'], row[q]), axis=1
        )

    return df_final

if __name__ == "__main__":
    results = run_lpp_process('final_100_sku_with_PN.csv', 'Price List Table.yaml')
    
    
    print(results[['SKU', 'MAP', 'Q1', 'LPP_Q1','Q2' ,'LPP_Q2','Q3','LPP_Q3','Q4', 'LPP_Q4']].head(12))
    

    results.to_csv('calculated_lpp_output.csv', index=False)
```

```python
import pandas as pd
import warnings
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os


load_dotenv()

user = os.getenv("DB_USER")
password = os.getenv("DB_PASSWORD")
host = os.getenv("DB_HOST")
port = os.getenv("DB_PORT")
db_name = os.getenv("DB_NAME")


connection_uri = f"mysql+pymysql://{user}:{password}@{host}:{port}/{db_name}"

warnings.filterwarnings('ignore')

try:

    engine = create_engine(connection_uri)
    
    
    PLTABLE = pd.read_json("pl table.json", lines=True) 


    PLTABLE['sub_category'] = PLTABLE['PL'].str.split('_').str[-1]

    PLTABLE

    PLTABLE.to_sql('PLTABLE', engine, index=False, if_exists='replace')
    print("Success: Data uploaded to MySQL!")

except Exception as e:
    print(f"An error occurred: {e}")

finally:

    engine.dispose()
```

```python
import pandas as pd
from docx import Document

price_df = pd.read_csv("price_monitoring_table.csv")
violation_df = pd.read_csv("violation_table.csv")

merged_df = pd.merge(price_df, violation_df, left_on="SKU", right_on="sku", how="inner")

merged_df['LPP_numeric'] = merged_df["LPP"].str.extract(r"([\d.]+)").astype(float)
merged_df['Advertised_price'] = pd.to_numeric(merged_df['Advertised_price'], errors='coerce')
merged_df['MAP_Price'] = pd.to_numeric(merged_df['MAP_Price'], errors='coerce')

violations = merged_df[
    (merged_df["Advertised_price"] < merged_df["LPP_numeric"]) |
    (merged_df["Advertised_price"] > merged_df["MAP_Price"])
]

def create_violation_letter(row):
    doc = Document()
    doc.add_heading("Violation Notice", level=1)

    
    content = (
        f"Violation Notice\n"
        f"Seller Name: {row['seller_name']}\n"
        f"Marketplace: {row['Marketplace']}\n"
        f"Category: {row['Category']} - {row['Sub_category']}\n"
        f"Advertised Price: {row['Advertised_price']}\n"
        f"LPP Price: {row['LPP']}\n"
        f"MAP Price: {row['MAP_Price']}\n"
        f"\nThis is to notify you that your advertised price has violated "
        f"the pricing policy. Please ensure compliance with the LPP and MAP guidelines.\n"
        f"{'-'*30}"
    )


    for line in content.split('\n'):
        doc.add_paragraph(line)

    
    filename = f"Violation_{str(row['seller_name']).replace(' ', '_')}_{row['SKU']}.docx"
    doc.save(filename)
    
    return content

print(f"Found {len(violations)} violations. Generating letters...\n")

for _, row in violations.iterrows():
    letter_text = create_violation_letter(row)
    print(letter_text) 
```
